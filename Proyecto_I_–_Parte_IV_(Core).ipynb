```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/retail_sales_dataset.csv')

# Calculate and display descriptive statistics
descriptive_stats = df.describe()
display(descriptive_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for visualization
numerical_cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']

# Create histograms for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Create boxplots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Resample data by day and calculate the sum of 'Total Amount'
daily_sales = df['Total Amount'].resample('D').sum()

# Create a line plot of daily sales trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_sales)
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for scatter plots
numerical_cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']

# Create scatter plots for pairs of numerical columns
sns.pairplot(df[numerical_cols])
plt.suptitle('Scatter Plots of Numerical Variables', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Create the histogram subplot
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.histplot(df['Total Amount'], kde=True)
plt.title('Histogram of Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')

# Create the boxplot subplot
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.boxplot(y=df['Total Amount'])
plt.title('Boxplot of Total Amount')
plt.ylabel('Total Amount')

plt.tight_layout()
plt.show()
```
